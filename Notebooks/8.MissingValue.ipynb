```python
from sklearn.impute import MissingIndicator
import pandas as pd

indicator = MissingIndicator(features="missing-only")

missing_features = indicator.fit_transform(df)

print(missing_features.shape)

missing_df = pd.DataFrame(
    missing_features,
    columns=df.columns[indicator.features_]
)

missing_df.head()
```

```python
df_indicator = pd.concat(
    [df, missing_df],
    axis=1
)
```
