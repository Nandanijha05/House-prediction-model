```python
import pandas as pd
```

```python
df = pd.read_csv("../data/train.csv")
```

```python
df["LotArea"].describe()
```

```python
df["LotArea_Bin"] = pd.cut(
    df["LotArea"],
    bins=4
)
```

```python
df[["LotArea", "LotArea_Bin"]].tail()
```

```python
df["LotArea_Bin"] = pd.cut(
    df["LotArea"],
    bins=4,
    labels=["Small", "Medium", "Large", "Very Large"]
)
```

```python
df[["LotArea", "LotArea_Bin"]].tail()
```

```python
df["LotArea_QBin"] = pd.qcut(
    df["LotArea"],
    q=4
)
```

```python
from sklearn.preprocessing import Binarizer
```

```python
binarizer = Binarizer(threshold=1)
```

```python
df["GarageCars_Binary"] = binarizer.fit_transform(
    df[["GarageCars"]]
)
```

```python
df["GrLivArea_bin"]=pd.cut(
    df["GrLivArea"],
    bins=4,
    labels=["Small","Medium","Large","Very Large"]

)
```

```python
df["GrLivArea_bin"].tail()
```

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
kbin = KBinsDiscretizer(
    n_bins=4,
    encode="ordinal",
    strategy="uniform"
)
```

```python
df["GrLivArea_KBin"] = kbin.fit_transform(
    df[["GrLivArea"]]
)
```
