```python
import pandas as pd
```

```python
df = pd.read_csv("../data/train.csv")
```

```python
df_random = df.copy()
```

```python
df_random["LotFrontage"].isnull().sum()
```

```python
non_missing = df_random["LotFrontage"].dropna()
```

```python
df_random.loc[
    df_random["LotFrontage"].isnull(),
    "LotFrontage"
] = non_missing.sample(
    df_random["LotFrontage"].isnull().sum(),
    random_state=42
).values
```

```python
df_random["LotFrontage"].isnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

df["LotFrontage"].plot(kind="kde", label="Original",color='red')

df_random["LotFrontage"].plot(kind="kde", label="Random Sample", color='green')

plt.legend()
plt.show()
```
